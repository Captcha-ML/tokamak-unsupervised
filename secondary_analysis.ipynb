```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans  # or another clustering method if preferred

# Assuming all_df is your concatenated dataframe
all_df['time'] = all_df['time']
unique_times = all_df['time'].drop_duplicates().sort_values().values

# Dictionary to store results
results = {}
time_list = np.concatenate([unique_times[:1000], unique_times[23000:]])
for time in time_list:
    # Filter data for the current time
    current_df = all_df[all_df['time'] == time]

    # Assuming X is a subset of columns used for PCA
    X = current_df.drop(['time','LHD_label'], axis=1)
    y = current_df['LHD_label']
    # Standardize the data
    scaler = StandardScaler()
    X_standardized = scaler.fit_transform(X)
    try:
        # Perform PCA
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X_standardized)
        cluster = KMeans(n_clusters=3) 
        cluster_labels = cluster.fit_predict(X_pca)
    
        # Store results
        results[time] = {
            'PCA_components': X_pca,
            'clusters': cluster_labels,
            'label': y
        }
    except:
        print(f"Failed PCA for time {time}")
```

```python
## import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

%matplotlib inline

def plot_pca_results(results, include_clusters=False):
    """
    Plot PCA results with an interactive time slider.

    :param results: Dictionary containing PCA (and optionally clustering) results for each time value.
    :param include_clusters: Boolean indicating whether to color points based on clustering labels.
    """
    # Create an output widget to display plot
    plot_output = widgets.Output()

    def update_plot(time):
        data = results[time]
        pca_components = data['PCA_components']

        plt.ion()
        plt.figure(figsize=(10, 6))
            # Scatter plot of the first two PCA components
        plt.scatter(pca_components[:, 0], pca_components[:, 1],
                    c=data['clusters'] if include_clusters else 'blue',
                    cmap='viridis', alpha=0.5)

        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.title(f'PCA for time = {time}')
        plt.grid(True)
        plt.tight_layout()
        
        plt.show()

    # Creating a slider for time selection
    time_slider = widgets.SelectionSlider(
        options=list(results.keys()),
        value=list(results.keys())[0],
        description='Time',
        disabled=False,
        continuous_update=False,
        orientation='horizontal',
        readout=True
    )

    # Link the widget to the update_plot function
    interactive_plot = widgets.interactive_output(update_plot, {'time': time_slider})
    display(time_slider, interactive_plot)


plot_pca_results(results, include_clusters=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display
from matplotlib.lines import Line2D
from matplotlib.colors import Normalize

%matplotlib inline

def plot_pca_results(results):
    """
    Plot PCA results with two interactive subplots: one for true labels and one for clustering labels,
    including dynamic legends for label colors.

    :param results: Dictionary containing PCA results and clustering results for each time value.
    """

    def update_plot(time):
        data = results[time]
        pca_components = data['PCA_components']
        true_labels = data['label']
        cluster_labels = data['clusters']

        plt.ioff()  # Turn off interactive mode
        fig, axs = plt.subplots(1, 2, figsize=(15, 6))  # 1 row, 2 columns

        # Normalizers for color mapping
        norm_true = Normalize(vmin=min(true_labels), vmax=max(true_labels))
        norm_cluster = Normalize(vmin=min(cluster_labels), vmax=max(cluster_labels))

        # Subplot for true labels
        scatter_true = axs[0].scatter(pca_components[:, 0], pca_components[:, 1], 
                                      c=true_labels, cmap='viridis', alpha=0.5, norm=norm_true)
        axs[0].set_xlabel('PCA Component 1')
        axs[0].set_ylabel('PCA Component 2')
        axs[0].set_title(f'True Labels for time = {time}')

        # Subplot for cluster labels
        scatter_cluster = axs[1].scatter(pca_components[:, 0], pca_components[:, 1], 
                                         c=cluster_labels, cmap='Accent', alpha=0.5, norm=norm_cluster)
        axs[1].set_xlabel('PCA Component 1')
        axs[1].set_ylabel('PCA Component 2')
        axs[1].set_title(f'Cluster Labels for time = {time}')

        # Create and update legends
        # For true labels
        unique_labels_true = np.unique(true_labels)
        legend_elements_true = [Line2D([0], [0], marker='o', color='w', label=f'Label {label}',
                                       markerfacecolor=plt.cm.viridis(norm_true(label)), markersize=10)
                                for label in unique_labels_true]
        axs[0].legend(handles=legend_elements_true, title='True Labels')

        # For cluster labels
        unique_labels_cluster = np.unique(cluster_labels)
        legend_elements_cluster = [Line2D([0], [0], marker='o', color='w', label=f'Label {label}',
                                          markerfacecolor=plt.cm.Accent(norm_cluster(label)), markersize=10)
                                   for label in unique_labels_cluster]
        axs[1].legend(handles=legend_elements_cluster, title='Cluster Labels')

        plt.tight_layout()
        plt.show()

    # Creating a slider for time selection
    time_slider = widgets.SelectionSlider(
        options=list(results.keys()),
        value=list(results.keys())[0],
        description='Time',
        disabled=False,
        continuous_update=False,
        orientation='horizontal',
        readout=True
    )

    # Link the widget to the update_plot function
    interactive_plot = widgets.interactive(update_plot, time=time_slider)
    display(interactive_plot)
```
